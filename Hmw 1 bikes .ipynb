```python
import pandas as pd
```

```python
data = pd.read_excel("bike.xlsx", sheetname = 2)
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(5)
```

```python
inputs = data.iloc[:,1:]
```

```python
inputs.head()
```

```python
data["claster"] = cluster.fit_predict(inputs)
```

```python
data.head()
```

```python
data["claster"].value_counts()
```

```python
new_data = data.iloc[:,[0,-1]]
new_data.head()
```

```python
new_data.to_excel("bike_clusters.xlsx")
```
